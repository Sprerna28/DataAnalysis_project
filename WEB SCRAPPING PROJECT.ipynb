# Web Scrapping of flipkart Website using BS4:-

```python
import bs4
import requests
```

```python
url="https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
data=requests.get(url)
data
```

```python
soup=bs4.BeautifulSoup(data.text,'html.parser')
print(soup.prettify())
```

```python
laptops=soup.find_all('div',{'class':'_4rR01T'})
laptops
```

```python
len(laptops)# length of laptops
```

```python
name=soup.find_all('div',{'class':'_4rR01T'})# div name taken form the live website
#print(name[0].text)
```

```python
name=soup.find_all('div',{'class':'_4rR01T'})# div name taken form the live website
#print(name[0].text)
#len(name)
```

```python
price=soup.find_all('div',{'class':'_30jeq3 _1_WHN1'})
#print(price[0].text)
```

```python
price=soup.find_all('div',{'class':'_30jeq3 _1_WHN1'})
#len(name)
```

```python
ratings=soup.find_all('div',{'class':'_3LWZlK'})
#print(ratings[0].text)
len(ratings)
```

```python
# we have to store all feched data from the live website into a csv file
```

```python
nm=[]
pr=[]
rt=[]
for i in name:
    nm.append(i.text)
    

for i in price:
    pr.append(i.text)
    
    
for i in range(len(nm)):
    rt.append(ratings[i].text)
    
```

```python
data={'Name':nm,'Ratings':rt,'Price':pr}
import pandas as pd
df=pd.DataFrame(data)
df.head()
```

```python
df.to_csv('laptops.csv',index=False)# csv file don't accsess formatting
```
